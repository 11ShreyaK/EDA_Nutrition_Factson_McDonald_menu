<h1>2.Now we Find out average Calories in each category in McDonald menu</h1>

```python
average_calories_per_category = ds.groupby('Category')['Calories'].mean().reset_index().sort_values(by='Calories', ascending=False)
average_calories_per_category
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Calories', y='Category', data=average_calories_per_category, palette='viridis',hue='Category')

plt.xlabel('Average Calories')
plt.ylabel('Category')
plt.title('Average Calories in Each Category')
plt.show()
```

<h5>Chicken & Fish Category has highest contribution to Calories and Beverages has lowest</h5>

<h1>3.Nutrient Distribution(carbohydrates, proteins, fats) in each Gategory</h1>

```python
nutrient_means_per_category = ds.groupby('Category')[['Carbohydrates', 'Protein', 'Total Fat']].mean().sort_values(by=['Carbohydrates','Protein','Total Fat'],ascending=[False, False, False])

nutrient_means_per_category.plot(kind='bar', figsize=(12, 8), colormap='viridis')
plt.xlabel('Category')
plt.ylabel('Nutrient Value (g)')
plt.title('Nutrient Distribution in Each Category')
plt.xticks(rotation=45, ha='right')
plt.show()
```

<h5>Breakfast Items</h5>

```python
breakfast_df = ds[ds['Category'] == 'Breakfast']
top_items = breakfast_df.sort_values(by=['Carbohydrates','Total Fat','Protein'], ascending=[False, False,False]).head(10)
plt.figure(figsize=(12, 8))
plt.barh(top_items['Item'], top_items['Carbohydrates'], color='blue', label='Carbohydrates')
plt.barh(top_items['Item'], top_items['Total Fat'], color='orange', label='Total Fat')
plt.barh(top_items['Item'], top_items['Protein'], color='green', label='Protein')

plt.xlabel('Nutrient Value')
plt.ylabel('Item')
plt.title('Top 10 Breakfast Items with Highest Proteins and Carbohydrates')
plt.legend()
plt.show()
```

<h1>4.Total Sugar Content by Category</h1>

```python
category_sugar_sum = ds.groupby('Category')['Sugars'].sum()
plt.figure(figsize=(12, 8))
category_sugar_sum.sort_values(ascending=False).plot(kind='bar', color='skyblue')

plt.xlabel('Category')
plt.ylabel('Total Sugar Content')
plt.title('Total Sugar Content by Category')
plt.show()
```

<h5> Coffee& Tea has highest Sugar content and Salads has low</h5>
<br>


<h5>Sugar content in salad with their items</h5>

<h1>5.Sodium Contribution by Each category</h1>

```python
category_sodium_sum = ds.groupby('Category')['Sodium'].sum()
category_sodium_sum_sorted = category_sodium_sum.sort_values(ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x=category_sodium_sum_sorted.values, y=category_sodium_sum_sorted.index)
plt.xlabel('Total Sodium Content')
plt.ylabel('Category')
plt.title('Total Sodium Contribution by Category') 
plt.show()
```

<h1>6.Iron and Dietary Fiber</h1>

```python
category_nutrients_sum = ds.groupby('Category')[['Iron (% Daily Value)', 'Dietary Fiber (% Daily Value)']].sum()
category_nutrients_sum = category_nutrients_sum.reset_index().melt(id_vars='Category', var_name='Nutrient', value_name='Total % Daily Value')

plt.figure(figsize=(12, 8))
sns.barplot(x='Total % Daily Value', y='Category', hue='Nutrient', data=category_nutrients_sum, palette='viridis')
plt.xlabel('Total % Daily Value')
plt.ylabel('Category')
plt.title('Total % Daily Value Contribution by Category')
plt.legend(title='Nutrient', loc='upper right', bbox_to_anchor=(1.2, 1))
plt.show()
```
